# Calculating Similarity

```python
from numpy import dot
from numpy.linalg import norm

data = user_interest.values
users = user_interest.index
m, k = data.shape

mat = np.zeros((m, m))

for i in range(m):
    for j in range(m):
        if i != j:
            mat[i][j] = dot(data[i], data[j])/(norm(data[i])*norm(data[j]))
        else:
            mat[i][j] = 0.
```

```python
similarity_matrix = pd.DataFrame(mat, columns=users, index=users)
similarity_matrix.iloc[:5, :5]
```

# Sorting

```python
class Posts:
    def __init__(self, postDF, sim_dict):
        self.postDF = postDF
        self.sim_dict = sim_dict
    
    def reorder_ranks(self, rank_col):

        previous_rank = rank_col[0] 
        new_ranks = [previous_rank]

        for rank in rank_col[1:]:
            if rank - previous_rank == 0:
                new_ranks.append(new_ranks[-1]) 
                previous_rank = rank
            elif rank - previous_rank > 0:
                new_ranks.append(new_ranks[-1]+1)
                previous_rank = rank
        return new_ranks
    
    def sort_posts(self, user, by_recency=False, by_thread=False, by_reply_numbers=False):
        
        cols_to_display = ["uid", "post_time", "text", "hashtags", "post_id", "parent_id","reply_numbers", "score", "rank"]
        
        df = pd.merge(self.postDF, pd.Series(self.sim_dict[user], name="score"), left_on="uid", right_index=True)
        df.sort_values(['score', 'post_time'], ascending=False, inplace=True)   
        df["rank"] = list(range(1, df.shape[0]+1))
        
        if by_thread:
            cols_to_keep3 = df.columns.drop(["parent_id", "score", "rank"])
            
            df_with_parent_ids = df[(df["parent_id"]!='0')]
            df_without_parent_ids = df[(df["parent_id"]=='0')]
            
            parent_id_lists = df_with_parent_ids["parent_id"].unique()
            
            post_parent_df = pd.DataFrame({"parent_id": []})
            post_parent_df["post_id"] = df_with_parent_ids["parent_id"].drop_duplicates()
            post_parent_df["parent_id"] = df_with_parent_ids["parent_id"].drop_duplicates()
            
            df_with_parent_ids = pd.concat([post_parent_df, df_with_parent_ids]).sort_values("parent_id")
            
            df_1 = pd.merge(df[["parent_id", "score", "rank"]].drop_duplicates(["parent_id"], keep="first"), df_with_parent_ids[["post_id", "parent_id"]], left_on="parent_id", right_on="parent_id")
            df_2 = pd.merge(df_1[["post_id"]], df[cols_to_keep3], left_on="post_id", right_on="post_id").drop_duplicates()
            df_3 = pd.merge(df_1, df_2, left_on="post_id", right_on="post_id")
            
            df = pd.concat([df_3, df_without_parent_ids]).sort_values("rank")
            
            df = df[~(df["post_id"].isin(parent_id_lists) & (df["parent_id"]=='0'))].reset_index(drop=True)
            
            df["rank"] = self.reorder_ranks(df["rank"].to_list())
                            
        if (by_recency) & (not by_thread):
            df = df.sort_values('rank', ascending=True) \
                    .groupby(['score'], sort=False) \
                    .apply(lambda x: x.sort_values(['post_time', 'reply_numbers'], ascending=False)) \
                    .reset_index(drop=True)
            
        elif (by_reply_numbers) & (not by_thread):
            df = df.sort_values('score', ascending=False) \
                    .groupby(['score'], sort=False) \
                    .apply(lambda x: x.sort_values(['reply_numbers', 'post_time'], ascending=False)) \
                    .reset_index(drop=True)
            df["rank"] = list(range(1, df.shape[0]+1))
            
        elif (by_reply_numbers & by_thread):
            df = df.sort_values('score', ascending=False) \
                    .groupby(['score'], sort=False) \
                    .apply(lambda x: x.sort_values(['reply_numbers', 'post_time'], ascending=False)) \
                    .reset_index(drop=True)
            df["rank"] = list(range(1, df.shape[0]+1))
            
        elif (by_recency & by_thread):
            df = df.sort_values('rank', ascending=True) \
                    .groupby(['score'], sort=False) \
                    .apply(lambda x: x.sort_values(['post_time', 'reply_numbers'], ascending=False)) \
                    .reset_index(drop=True)
            df["rank"] = list(range(1, df.shape[0]+1))

        
        
        return df[cols_to_display]
```

```python
similirty_dict = similarity_matrix.to_dict('dict')
```

```python
p = Posts(posts_df, similirty_dict)
# "e31e7cf3-9c4b-4da7-be33-db0e0cf4edd7"

df_sorted = p.sort_posts('0ba4476e-a542-470c-a41d-2a8c6a45ddc1', by_thread=True, by_recency=False, by_reply_numbers=True)
# df_test.sort_values(["ordered_rank", "post_time"], ascending=[True, False])
df_sorted
```
